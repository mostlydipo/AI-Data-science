```python
import numpy as np # Mathematics computation
import matplotlib.pyplot as plt  # Plots and graphs
import pandas as pd # Data processing(files)

# Sklearn library for machine learning
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
nba=pd.read_csv('nba_rookie_data.csv') # Extract data.
nba.head() # First 5 rows of the data.
```

**Data cleaning(Remove errorneous values)**

```python
nba.isnull().any().any()  # Check for null values.
```

```python
nba.dropna(inplace= True) # Drop null values where null is True
```

```python
nb = nba.drop('Name', axis=1) # Drop name column.
nb
```

**Data wrangling(Scaling and structruting of data)**

```python
scaler = MinMaxScaler() # Transforms numeric feature data to range 0 to 1.
 
nbss = scaler.fit_transform(nb.to_numpy())
nbss = pd.DataFrame(nbss, columns=nb.columns)
 
print("Scaled Dataset Using MinMaxScaler")
nbss.head()
```

```python
x= nbss.iloc[:,0:19].values # Dependent variable
y= nbss.iloc[:,19].values # Independent variable
```

```python
x_train, x_test,y_train,y_test= train_test_split(x,y, test_size=1/3, random_state=0) # Train, test and split data.
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(40,10), activation= 'tanh', #Neural network
    random_state=0, max_iter=2000)
```

```python
mlp.fit(x_train,y_train)
```

```python
mlp.score(x_test,y_test) #The neural network model using Mlpclassifier is 72 percent accurate
```
