```python
import pandas as pd # Data processing(files)
import numpy as np # Mathematics computation
import matplotlib.pyplot as plt # Plots and graphs
```

```python
# Sklearn library for machine learning
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics

from sklearn.preprocessing import LabelEncoder
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
nba=pd.read_csv('nba_rookie_data.csv') # Extract data.
nba.head() # First 5 rows of the data.
```

**Data cleaning(Remove errorneous values)**

```python
nba.isnull().any().any() # Check for null values.
```

```python
nba.columns[nba.isna().any()] # Check the column where null values exists
```

```python
nba.dropna(inplace= True) # Drop null values.
```

```python
nba.isna().any() # check if null value has been dropped
```

```python
nb = nba.drop('Name', axis=1) # Drop name column.
nb
```

**Data wrangling(Scaling and structruting of data)**

```python
scaler = MinMaxScaler() # Transforms numeric feature data to range 0 to 1.
 
nbs = scaler.fit_transform(nb.to_numpy())
nbs = pd.DataFrame(nbs, columns=[
    'Games Played',	'Minutes Played',	'Points Per Game',	'Field Goals Made',	'Field Goal Attempts',	'Field Goal Percent',	'3 Point Made',	'3 Point Attempt'	,'3 Point Percent',
    'Free Throw Made',	'Free Throw Attempts',	'Free Throw Percent',	'Offensive Rebounds',	'Defensive Rebounds',	'Rebounds',	'Assists',	'Steals', 'Blocks',	'Turnovers',	'TARGET_5Yrs'])
 
print("Scaled Dataset Using MinMaxScaler")
nbs.head()
```

```python
inputs= nbs.drop('TARGET_5Yrs', axis='columns') # Dependent variables.
target= nbs.TARGET_5Yrs # Indepenent variable.
```

```python
X_train, X_test,y_train,y_test= train_test_split(inputs,target, test_size=1/3, random_state=0) # Train, test and split data.
```

```python
len(X_train) # Number of traing data
```

```python
len(X_test) # Number of testing data
```

```python
len(inputs) # Total number of data
```

**Naive Bayes models(Probalistic machine learning algorithm).**

```python
X_test[:5]
```

```python
gnb=GaussianNB() # 66 percent accurate on a player lasting 5 years in the NBA.
gnb.fit(X_train,y_train)
print(gnb)

gnb.score(X_test,y_test)
```

```python
y_test[:5]
```

```python
gnb.predict(X_test[:5]) # Our GaussianNB got 2 predictions wrong(1287,458) from our y_test above. consider we have 66 percent accuracy.
```

```python
bernnb = BernoulliNB(binarize = 0.1) # Best  Naive Bayes probalistic model on a player lasting 5 years in the NBA.
bernnb.fit(X_train, y_train)
print(bernnb)

bernnb.score(X_test, y_test)
```

```python
y_test[:5]
```

```python
bernnb.predict(X_test[:5]) # The BernoulliNB got 1 prediction wrong(1287) from our y_test. Consider we have 68 percent accuracy
```
